```python
df_ventas['TipoVenta'] = np.where(df_ventas['MontoFactura'] > 700, 'Alta', 'Normal')
df_ventas.head()
```

```python
df_ventas.set_index('NroFactura')
```

```python
df_ventas_alta = df_ventas[df_ventas['TipoVenta'] == 'Alta']
```
