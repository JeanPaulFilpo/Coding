# Ejercicio

- De df_ventas seleccionelos clientes tipo empresa que hayan generado un montofactura mayor o igual a 850 y guardelo en un dataframe llamado: df_ventas_empresas_grande

- Del nuevo dataframe df_ventas_empresas_grande convierta el campo NroFactura en el indice

- Del nuevo dataframe df_ventas_empresas_grande solo mantenga las columnas: Cliente, FechaFactura, Costo y MontoFactura. 

- Agregue una nueva columna que se llame porc_costo que sea el resultado de dividir la columna Costo entre MontoFactura

- Elimine la columna Costo

- Cambie el nombre a la columna Cliente y pongale: Nombre

- Guarde ese dataframe como un archivo de excel en el escritorio de su computadora con el nombre de: Empresas Grandes.xlsx.

```python
df_ventas_empresas_grande = df_ventas[(df_ventas['Tipo']=='Empresa') & (df_ventas['MontoFactura']>=850)]
```

```python
df_ventas_empresas_grande.set_index('NroFactura', inplace = True)
```

```python
df_ventas_empresas_grande = df_ventas_empresas_grande[['Cliente','FechaFactura','Costo','MontoFactura']]
```

```python
df_ventas_empresas_grande['porc_costo'] = df_ventas_empresas_grande['Costo']/df_ventas_empresas_grande['MontoFactura']
```

```python
df_ventas_empresas_grande.drop(columns=['Costo'], inplace=True)
```

```python
df_ventas_empresas_grande.rename(columns={'Cliente':'Nombre'})
```

```python
df_ventas_empresas_grande.to_excel(r'C:\Users\gilbe\OneDrive\Desktop\Empresas Grandes.xlsx')
```
