# An√°lisis Exploratorio de Datos

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
propinas = sns.load_dataset('tips')
```

```python
propinas.head()
```

# Matplotlib

```python
#Gr√°fico de Barra
plt.bar(propinas['sex'],propinas['total_bill'])
plt.xlabel('Sexo')
plt.ylabel('Total de la Cuenta')
plt.title('Propinas')
plt.show()
```

```python
plt.hist(propinas[propinas['sex']=='Male']['tip'],
         bins = 10, 
         alpha = 0.2, 
         color = 'blue',
         label ='Hombres')

plt.hist(propinas[propinas['sex']=='Female']['tip'],
         bins = 10, 
         alpha = 0.2, 
         color = 'pink',
         label ='Mujeres')

plt.axvline(propinas[propinas['sex']=='Male']['tip'].mean(), label = 'Avg.Male', color = 'blue')
plt.axvline(propinas[propinas['sex']=='Female']['tip'].mean(), label = 'Avg.Female', color = 'red')

plt.legend()
plt.show()
```

```python
plt.scatter(propinas['tip'], propinas['total_bill'])
plt.show()
```

```python
#Gr√°fico de linea
plt.plot(propinas.index,propinas['total_bill'])
plt.xlabel('Ordenes')
plt.ylabel('Total de la Cuenta')
plt.title('Propinas')
plt.show()
```

```python
plt.boxplot(propinas['tip'])
plt.show()
```

```python
propinas_time = propinas.groupby('time').sum()['total_bill']
propinas_time
```

```python
#Gr√°fico de pastel
plt.pie(x = propinas_time,
        labels = propinas_time.index, 
        autopct = '%.2f')
plt.show()
```

```python
#Gr√°fico de estad√≠sticos
plt.hist(propinas['tip'])
plt.show()
```

```python
#diagrama de dispersion 
plt.scatter(propinas['tip'],propinas['total_bill'])
plt.show()
```

```python
#diagrama de caja
plt.boxplot(propinas['tip'])
plt.show()
```

```python
#COMBINACION DE GRAFICOS EN UN SOLO
plt.hist(propinas[propinas['sex']=='Female']['total_bill'], 
         bins=20, 
         alpha=0.2, 
         color='red', 
         label='Total Bill - Females')
plt.hist(propinas[propinas['sex']=='Male']['total_bill'], bins=20, alpha=0.2, color='blue', label='Total Bill - Females')
plt.axvline(propinas[propinas['sex']=='Female']['total_bill'].mean(), label='Avg Female', color='red')
plt.axvline(propinas[propinas['sex']=='Male']['total_bill'].mean(), label='Avg Male', color='blue')
plt.suptitle('Distribuci√≥n de Total de la Cuenta por Sexo')
plt.legend()
plt.show()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# Solo cuadr√≠culas
sns.PairGrid(iris)
```

```python
# Luego mapea a la cuadr√≠cula
g = sns.PairGrid(iris)
g.map(plt.scatter) #dibuja un scatter plot para todas las combinaciones de numeros
```

```python
# Mapear las partes superior, inferior y diagonal
g = sns.PairGrid(iris) #creo la cuadricula
g.map_diag(plt.hist) #histograma para las cuadriculas diagonales
g.map_upper(plt.scatter) #dispersion para las cuadriculas superiores que no son parte de la diagonal
g.map_lower(sns.kdeplot) #densidad para las cuadriculas inferiores que no son parte de la diagonal
```

```python
# Solo la cuadriculas
g = sns.FacetGrid(propinas, col="time", row="smoker")
```

```python
g = sns.FacetGrid(propinas, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(propinas, col="time",  row="smoker",hue='sex')
# Observe que los argumentos vienen despu√©s de la llamada plt.scatter
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```
